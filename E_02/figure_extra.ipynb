```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
from matplotlib.pyplot import ylabel, xlabel

plt.rcParams["text.usetex"] = True

## 1) initialiser le graphique
fig, ax = plt.subplots(2,2, figsize=(10,10))

## 2) calcul des variables au fil du temps pour premier ensemble de cond. initiales
# valeurs des constantes
sigma = 10
rho = 28
beta = 2.667
eta = np.sqrt(beta*(rho-1)) # ~8.5
epsilon = 0.01
n_pas_de_temps = 4e4
pas_de_temps = 0.001

# calculer valeurs initiales des variables dépendantes
x_0 = eta + 91
y_0 = eta + 3
z_0 = rho +90

# utilisation des équations discrétisées
valeurs_t = np.arange(0,pas_de_temps*n_pas_de_temps,pas_de_temps) # ndarray
valeurs_x = np.array([x_0]) # ndarray
valeurs_y = np.array([y_0])
valeurs_z = np.array([z_0])

for i in range(len(valeurs_t)-1):
    x_p_1 = (sigma*(valeurs_y[i]-valeurs_x[i]))*pas_de_temps+valeurs_x[i]
    y_p_1 = (valeurs_x[i]*(rho-valeurs_z[i])-valeurs_y[i])*pas_de_temps+valeurs_y[i]
    z_p_1 = (valeurs_y[i]*valeurs_x[i]-beta*valeurs_z[i])*pas_de_temps+valeurs_z[i]

    valeurs_x = np.append(valeurs_x,x_p_1)
    valeurs_y = np.append(valeurs_y,y_p_1)
    valeurs_z = np.append(valeurs_z,z_p_1)

# x selon z
ax[1,1].plot(valeurs_z,valeurs_x, color='k', linewidth=0.7)
ax[1,1].scatter(z_0, x_0, color="red", s=70, alpha=1) # color="orange", s=70
ax[1,1].set_title(r"$[x_0,y_0,z_0]=[\eta+91,\eta+3,\rho+90]$", fontsize=16)
ax[1,1].set_xlabel(r"$z$", fontsize=16)

## 3) calcul des variables au fil du temps pour second ensemble de cond. initiales
# calculer valeurs initiales des variables dépendantes
x_0 = eta + 6.5
y_0 = eta - 18.5
z_0 = rho - 13

# utilisation des équations discrétisées
valeurs_t = np.arange(0,pas_de_temps*n_pas_de_temps,pas_de_temps) # ndarray
valeurs_x = np.array([x_0]) # ndarray
valeurs_y = np.array([y_0])
valeurs_z = np.array([z_0])

for i in range(len(valeurs_t)-1):
    x_p_1 = (sigma*(valeurs_y[i]-valeurs_x[i]))*pas_de_temps+valeurs_x[i]
    y_p_1 = (valeurs_x[i]*(rho-valeurs_z[i])-valeurs_y[i])*pas_de_temps+valeurs_y[i]
    z_p_1 = (valeurs_y[i]*valeurs_x[i]-beta*valeurs_z[i])*pas_de_temps+valeurs_z[i]

    valeurs_x = np.append(valeurs_x,x_p_1)
    valeurs_y = np.append(valeurs_y,y_p_1)
    valeurs_z = np.append(valeurs_z,z_p_1)

# x selon z
ax[0,1].plot(valeurs_z,valeurs_x, color='k', linewidth=0.7)
ax[0,1].scatter(z_0, x_0, color="red", s=70, alpha=1)
ax[0,1].set_title(r"$[x_0,y_0,z_0]=[\eta+6.5,\eta-18.5,\rho-13]$", fontsize=16)

## 4) calcul des variables au fil du temps pour troisième ensemble de cond. initiales
# calculer valeurs initiales des variables dépendantes
x_0 = eta -29
y_0 = eta +8.5
z_0 = rho - 28

# utilisation des équations discrétisées
valeurs_t = np.arange(0,pas_de_temps*n_pas_de_temps,pas_de_temps) # ndarray
valeurs_x = np.array([x_0]) # ndarray
valeurs_y = np.array([y_0])
valeurs_z = np.array([z_0])

for i in range(len(valeurs_t)-1):
    x_p_1 = (sigma*(valeurs_y[i]-valeurs_x[i]))*pas_de_temps+valeurs_x[i]
    y_p_1 = (valeurs_x[i]*(rho-valeurs_z[i])-valeurs_y[i])*pas_de_temps+valeurs_y[i]
    z_p_1 = (valeurs_y[i]*valeurs_x[i]-beta*valeurs_z[i])*pas_de_temps+valeurs_z[i]

    valeurs_x = np.append(valeurs_x,x_p_1)
    valeurs_y = np.append(valeurs_y,y_p_1)
    valeurs_z = np.append(valeurs_z,z_p_1)

# x selon z
ax[1,0].plot(valeurs_z,valeurs_x, color='k', linewidth=0.7)
ax[1,0].scatter(z_0, x_0, color="red", s=70, alpha=1)
ax[1,0].set_title(r"$[x_0,y_0,z_0]=[\eta-29,\eta+8.5,\rho-28]$", fontsize=16)
ax[1,0].set_ylabel(r"$x$", fontsize=16)
ax[1,0].set_xlabel(r"$z$", fontsize=16)

## 5) calcul des variables au fil du temps pour cond. initiales utilisées souvent dans le travail
# calculer valeurs initiales des variables dépendantes
x_0 = eta
y_0 = eta + 3
z_0 = rho - 1

# utilisation des équations discrétisées
valeurs_t = np.arange(0,pas_de_temps*n_pas_de_temps,pas_de_temps) # ndarray
valeurs_x = np.array([x_0]) # ndarray
valeurs_y = np.array([y_0])
valeurs_z = np.array([z_0])

for i in range(len(valeurs_t)-1):
    x_p_1 = (sigma*(valeurs_y[i]-valeurs_x[i]))*pas_de_temps+valeurs_x[i]
    y_p_1 = (valeurs_x[i]*(rho-valeurs_z[i])-valeurs_y[i])*pas_de_temps+valeurs_y[i]
    z_p_1 = (valeurs_y[i]*valeurs_x[i]-beta*valeurs_z[i])*pas_de_temps+valeurs_z[i]

    valeurs_x = np.append(valeurs_x,x_p_1)
    valeurs_y = np.append(valeurs_y,y_p_1)
    valeurs_z = np.append(valeurs_z,z_p_1)

# x selon z
ax[0,0].plot(valeurs_z,valeurs_x, color='k', linewidth=0.7)
ax[0,0].scatter(z_0, x_0, color="red", s=70, alpha=1)
ax[0,0].set_title(r"$[x_0,y_0,z_0]=[\eta,\eta+3,\rho-1]$", fontsize=16)
ax[0,0].set_ylabel(r"$x$", fontsize=16)

plt.tight_layout()
plt.savefig('/Users/evecastonguay/Desktop/Labo/E_02/Figures/extra1.png',dpi=500)
plt.show()
```

```python
# imports
import numpy as np
import matplotlib.pyplot as plt
import random
from matplotlib.animation import FuncAnimation
from scipy.integrate import odeint
import matplotlib.animation as animation
plt.rcParams["text.usetex"] = True

fig = plt.figure(figsize=(8,9))
ax = fig.add_subplot(projection="3d")
#fig, ax = plt.subplots(subplot_kw={'projection':'3d'}, figsize=(8,9)) # dictionnaire avec keywords spécifiques pour créer graphe 3d
plt.subplots_adjust(left=0, bottom=0, right=1, top=1)

## 1) constantes
sigma = 10
rho = 28
beta = 2.667

## 2) cette fois-ci, utiliser odeint pour résoudre le système
# découvert grâce à https://github.com/Younes-Toumi/Youtube-Channel/blob/b3dca52cbb4ca831329a1efe795a9b4863ecda05/Simulation%20with%20Python/Lorenz%20Attractor/lorenz_attractor.ipynb et partiellement copié de.
def system_of_odes(vector, t, sigma, beta, rho): # définit le système d'équations différentielles à résoudre. t obligatoire car
    # fonction passée à odeint doit avoir forme (y, t, ...
    x, y, z = vector # valeurs à résoudre
    d_vector = [
        sigma*(y - x),
        x*(rho - z) - y,
        x*y - beta*z
    ]
    return d_vector # retourne le vecteur de dérivées

n_pas_de_temps = 2000 #4e4
pas_de_temps = 0.001
stop = n_pas_de_temps*pas_de_temps
time_points = np.linspace(0,stop,num=n_pas_de_temps)

# générer une trajectoire à partir d'une position initiale aléatoire
def trajectoire():
    "Retourne une trajectoire dans un array de taille (time_points,3)"
    position_0 = [random.randint(-100, 100) for i in range(3)]
    print(position_0)
    positions = odeint(system_of_odes, position_0, time_points, args=(sigma, beta, rho))
    return positions

# générer les n trajectoires
n = 10
trajectoires = [trajectoire() for index in range(n)]

# Définir les limites des axes une seule fois
all_points = np.vstack(trajectoires)
ax.set_xlim(np.min(all_points[:, 0]), np.max(all_points[:, 0]))
ax.set_ylim(np.min(all_points[:, 1]), np.max(all_points[:, 1]))
ax.set_zlim(np.min(all_points[:, 2]), np.max(all_points[:, 2]))

# nommer les axes
ax.set_xlabel(r"$x$", fontsize=14)
ax.set_ylabel(r"$y$", fontsize=14)
ax.set_zlabel(r"$z$", fontsize=14)

# générer les lignes vides
lignes = [ax.plot([], [], [])[0] for _ in trajectoires]

# fonction pour remplir les lignes
def update(num, trajectoires, lignes): # créer les «frames» une par une
    start=max((num-100,0)) # pour que la trajectoire s'efface graduellement
    for ligne, traj in zip(lignes, trajectoires):
        ligne.set_data_3d(traj[start:num, :].T) # mettre coordonnées dans cases vides de lignes
    return lignes # maintenant rempli

# animer (https://matplotlib.org/stable/users/explain/animations/animations.html)
ani = animation.FuncAnimation(fig, update, n_pas_de_temps, fargs=(trajectoires, lignes), interval=30, blit=False) # interval = delay between frames in ms

# enregistrer
ani.save("/Users/evecastonguay/Desktop/ani5.mp4", writer='ffmpeg', fps=200) # 90 semble optimal
plt.show()

# n_pas_de_temps = 4e4:
#stop = int(n_pas_de_temps*pas_de_temps): temps total (40s)
#time_points = np.linspace(0,stop,num=int(n_pas_de_temps)): liste des moments dans temps où valeurs sont calculées ->frames
```
