#### Imports

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from matplotlib.pyplot import subplot, ylabel, xlabel
import seaborn as sns

plt.rcParams["text.usetex"] = True
```

#### 1) Planche des solutions stationnaires, périodiques et chaotiques (pour x, y, z)

```python
## 0) constantes pour le problème
pas_de_temps = 0.001
n_pas_de_temps = 2e4

## 1) préparation de la planche
sns.set_style("darkgrid", {"axes.facecolor": "0.9"}) # controller le style avec seaborn: plus la valeur est basse, plus c'est foncé.
fig, ax = plt.subplots(3,1, figsize=(9, 9), sharex=True) # figsize=(8, 8), sharey=True

## 2.1) calcul des valeurs de x, y, z sur tous les pas de temps (stationnaire)
# valeurs des constantes
sigma = 10
rho = 1
beta = 2.667
eta = np.sqrt(beta*(rho-1))

# calculer valeurs initiales des variables dépendantes
x_0 = eta
y_0 = eta + 3
z_0 = rho - 1

# utilisation des équations discrétisées
valeurs_t = np.arange(0,pas_de_temps*n_pas_de_temps,pas_de_temps) # ndarray
valeurs_x = np.array([x_0]) # ndarray
valeurs_y = np.array([y_0])
valeurs_z = np.array([z_0])

for i in range(len(valeurs_t)-1):
    x_p_1 = (sigma*(valeurs_y[i]-valeurs_x[i]))*pas_de_temps+valeurs_x[i]
    y_p_1 = (valeurs_x[i]*(rho-valeurs_z[i])-valeurs_y[i])*pas_de_temps+valeurs_y[i]
    z_p_1 = (valeurs_y[i]*valeurs_x[i]-beta*valeurs_z[i])*pas_de_temps+valeurs_z[i]

    valeurs_x = np.append(valeurs_x,x_p_1)
    valeurs_y = np.append(valeurs_y,y_p_1)
    valeurs_z = np.append(valeurs_z,z_p_1)

# graphique
ax[0].plot(valeurs_t,valeurs_x, color='midnightblue', alpha=0.6, label='x')
ax[0].plot(valeurs_t,valeurs_y, color='orangered', alpha=0.6, label = 'y')
ax[0].plot(valeurs_t,valeurs_z, color='darkgreen', alpha=0.5, label = 'z')
#ax[0].set_title(r"a) Solution stationnaire")
ax[0].set_title("a) Solution stationnaire")
ax[0].legend()

## 2.2) calcul des valeurs de x, y, z sur tous les pas de temps (périodique)
# valeurs des constantes
sigma = 10
rho = 350
beta = 2.667
eta = np.sqrt(beta*(rho-1))

# calculer valeurs initiales des variables dépendantes
x_0 = eta
y_0 = eta + 3
z_0 = rho - 1

# utilisation des équations discrétisées
valeurs_t = np.arange(0,pas_de_temps*n_pas_de_temps,pas_de_temps) # ndarray
valeurs_x = np.array([x_0]) # ndarray
valeurs_y = np.array([y_0])
valeurs_z = np.array([z_0])

for i in range(len(valeurs_t)-1):
    x_p_1 = (sigma*(valeurs_y[i]-valeurs_x[i]))*pas_de_temps+valeurs_x[i]
    y_p_1 = (valeurs_x[i]*(rho-valeurs_z[i])-valeurs_y[i])*pas_de_temps+valeurs_y[i]
    z_p_1 = (valeurs_y[i]*valeurs_x[i]-beta*valeurs_z[i])*pas_de_temps+valeurs_z[i]

    valeurs_x = np.append(valeurs_x,x_p_1)
    valeurs_y = np.append(valeurs_y,y_p_1)
    valeurs_z = np.append(valeurs_z,z_p_1)

# graphique
ax[1].plot(valeurs_t,valeurs_x, color='midnightblue', alpha=0.6, label='x')
ax[1].plot(valeurs_t,valeurs_y, color='orangered', alpha=0.6, label = 'y')
ax[1].plot(valeurs_t,valeurs_z, color='darkgreen', alpha=0.5, label = 'z')
ax[1].set_title(r"b) Solution périodique")
ax[1].legend()

## 2.3) calcul des valeurs de x, y, z sur tous les pas de temps (chaotique)
# valeurs des constantes
sigma = 10
rho = 28
beta = 2.667
eta = np.sqrt(beta*(rho-1))

# calculer valeurs initiales des variables dépendantes
x_0 = eta
y_0 = eta + 3
z_0 = rho - 1

# utilisation des équations discrétisées
valeurs_t = np.arange(0,pas_de_temps*n_pas_de_temps,pas_de_temps) # ndarray
valeurs_x = np.array([x_0]) # ndarray
valeurs_y = np.array([y_0])
valeurs_z = np.array([z_0])

for i in range(len(valeurs_t)-1):
    x_p_1 = (sigma*(valeurs_y[i]-valeurs_x[i]))*pas_de_temps+valeurs_x[i]
    y_p_1 = (valeurs_x[i]*(rho-valeurs_z[i])-valeurs_y[i])*pas_de_temps+valeurs_y[i]
    z_p_1 = (valeurs_y[i]*valeurs_x[i]-beta*valeurs_z[i])*pas_de_temps+valeurs_z[i]

    valeurs_x = np.append(valeurs_x,x_p_1)
    valeurs_y = np.append(valeurs_y,y_p_1)
    valeurs_z = np.append(valeurs_z,z_p_1)

# graphique
ax[2].plot(valeurs_t,valeurs_x, color='midnightblue', alpha=0.6, label='x')
ax[2].plot(valeurs_t,valeurs_y, color='orangered', alpha=0.6, label = 'y')
ax[2].plot(valeurs_t,valeurs_z, color='darkgreen', alpha=0.5, label = 'z')
ax[2].set_title(r"c) Solution chaotique")
ax[2].legend()

## 3) titre des axes du graphique
fig.text(0.535, 0.01, 'Temps (s)', ha='center', fontsize='large')

plt.tight_layout(rect=[0.02, 0.02, 1, 1])
plt.savefig('/Users/evecastonguay/Desktop/Labo/E_02/Figures/planche_xyz.png',dpi=500)
plt.show()
```
