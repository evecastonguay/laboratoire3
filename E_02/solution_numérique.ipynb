# Partie I: solution numérique du système

#### Imports

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from matplotlib.pyplot import subplot, ylabel, xlabel
import seaborn as sns
```

#### 1.1) La solution stationnaire

```python
##:     nom de l'étape
#:      nom de la sous-étape & commentaires

## 1) valeurs des constantes
sigma = 10
rho = 1
beta = 2.667
pas_de_temps = 0.001
n_pas_de_temps = 2e4

## 2) calcul des valeurs de x, y, z sur tous les pas de temps
# calculer valeurs initiales des variables dépendantes
x_0 = np.sqrt(beta*(rho-1))
y_0 = x_0 + 3
z_0 = rho - 1

# utilisation des équations discrétisées
valeurs_t = np.arange(0,pas_de_temps*n_pas_de_temps,pas_de_temps) # ndarray
valeurs_x = np.array([x_0]) # ndarray
valeurs_y = np.array([y_0])
valeurs_z = np.array([z_0])

for i in range(len(valeurs_t)-1):
    x_p_1 = (sigma*(valeurs_y[i]-valeurs_x[i]))*pas_de_temps+valeurs_x[i]
    y_p_1 = (valeurs_x[i]*(rho-valeurs_z[i])-valeurs_y[i])*pas_de_temps+valeurs_y[i]
    z_p_1 = (valeurs_y[i]*valeurs_x[i]-beta*valeurs_z[i])*pas_de_temps+valeurs_z[i]

    valeurs_x = np.append(valeurs_x,x_p_1)
    valeurs_y = np.append(valeurs_y,y_p_1)
    valeurs_z = np.append(valeurs_z,z_p_1)

## 3) graphiques des variables dépendantes selon le temps
sns.set_style("darkgrid", {"axes.facecolor": "0.9"}) # controller le style avec seaborn: plus la valeur est basse, plus c'est foncé.
fig, ax = plt.subplots(figsize=(15, 5)) # figsize=(8, 8), sharey=True
ax.plot(valeurs_t,valeurs_x, color='orange', alpha=0.7, label='x')
ax.plot(valeurs_t,valeurs_y, label = 'y')
ax.plot(valeurs_t,valeurs_z, label = 'z')
ax.set(xlabel='Temps (s)', ylabel='Valeur des variables étudiées (sans unités)')
ax.legend()

plt.savefig('/Users/evecastonguay/Desktop/Labo/E_02/Figures/sol_stat.png')
plt.show()

## 4) graphique x selon y
fig, ax = plt.subplots()
ax.plot(valeurs_y,valeurs_x, color='darkseagreen')
ax.set(xlabel='Valeurs de y (sans unités)',ylabel='Valeurs de x (sans unités)')

plt.savefig('/Users/evecastonguay/Desktop/Labo/E_02/Figures/x_selon_y_stat.png')
plt.show()

## 5) graphique x selon z
fig, ax = plt.subplots()
ax.plot(valeurs_z,valeurs_x, color='darkslateblue',alpha=0.6)
ax.set(xlabel='Valeurs de z (sans unités)',ylabel='Valeurs de x (sans unités)')

plt.savefig('/Users/evecastonguay/Desktop/Labo/E_02/Figures/x_selon_z_stat.png')
plt.show()


```
